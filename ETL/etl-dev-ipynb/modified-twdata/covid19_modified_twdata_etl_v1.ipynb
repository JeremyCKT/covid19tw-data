```python
import pandas as pd
import seaborn as sns

df = pd.read_excel("校正回歸_210528.xlsx")
cols = df.columns.tolist()
cols[0] = "date"
df.columns = cols
df = df.set_index("date")
df.columns.name = "released_date"
df = df.stack("released_date").to_frame("n").reset_index()
df = df[df["n"] > 0]

grouped = df.groupby("date")
new_gdf_data = []

for idx, gdf in grouped:
    #print("~~~~~~")
    gdf["last_n"] = gdf["n"].shift(1).fillna(0)
    gdf["diff_n"] = gdf["n"] - gdf["last_n"]
    #print(gdf)
    new_gdf_data.append(gdf)

df = pd.concat(new_gdf_data)
df = df[["date", "released_date", "n", "diff_n"]]
df = df[df["diff_n"] > 0]

df.to_csv("covid19_modified_twdata_20210528.csv", index=False)
```
